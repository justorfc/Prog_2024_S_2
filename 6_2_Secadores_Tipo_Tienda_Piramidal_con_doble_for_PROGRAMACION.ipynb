# Secadores Solares por Convección Natural

Las dimensiones de un secador solar tipo tienda piramidal dependen de la altura del secador, mediante las siguientes relaciones:

$$
B = \frac{2*H}{\tan(\alpha)}
$$


$$
A = \sqrt{\frac{B^2}{2}+H^2}
$$

$$
C = \sqrt{(\frac{B}{2})^2+H^2}
$$

Donde:

+ H es la altura del secador
+ B es la base del secador
+ A es la arista del secador
+ C la apotema del secador
+ $\alpha$ es el ángulo interior, puede ser: **51.85, 43.0, 68.0**

Desarrolle un programa que calcule las dimensiones de un secador para diferentes alturas.

````python
# prompt: Escribe código de python que resuelva el ejercicio: "Secadores Solares por Convección Natural"

import math

def calcular_dimensiones(altura, angulo):
  """
  Calcula las dimensiones de un secador solar tipo tienda piramidal.

  Args:
    altura: La altura del secador.
    angulo: El ángulo interior del secador.

  Returns:
    Una tupla con las dimensiones del secador: (base, arista, apotema).
  """
  angulo_rad = math.radians(angulo)
  base = (2 * altura) / math.tan(angulo_rad)
  arista = math.sqrt((base**2) / 2 + altura**2)
  apotema = math.sqrt((base / 2)**2 + altura**2)
  return (base, arista, apotema)

def generar_tabla(angulos):
  """
  Genera una tabla con las dimensiones del secador para diferentes alturas.

  Args:
    angulos: Una lista de ángulos interiores.
  """
  alturas = [0.5, 0.6, 0.7, 0.8, 0.9, 1.0, 1.1, 1.2, 1.3, 1.4, 1.5]
  for angulo in angulos:
    print(f"Para α = {angulo}")
    print("|ALTURA|BASE|ARISTA|APOTEMA|")
    print("|------|----|------|-------|")
    for altura in alturas:
      base, arista, apotema = calcular_dimensiones(altura, angulo)
      print(f"|{altura:.1f}|{base:.2f}|{arista:.2f}|{apotema:.2f}|")
    print("\n")

# Ángulos interiores a considerar
angulos = [51.85, 43.0, 68.0]

# Generar la tabla de dimensiones
generar_tabla(angulos)
````

````python
import math

# Función para calcular las dimensiones del secador solar
def calcular_dimensiones(H, alpha):
    # Convertir ángulo de grados a radianes para usar en las funciones trigonométricas
    alpha_radianes = math.radians(alpha)

    # Calcular la base B
    B = (2 * H) / math.tan(alpha_radianes)

    # Calcular la arista A
    A = math.sqrt((B ** 2) / 2 + H ** 2)

    # Calcular la apotema C
    C = math.sqrt((B / 2) ** 2 + H ** 2)

    return B, A, C

# Función para mostrar los resultados en una tabla
def mostrar_tabla(alpha, alturas):
    print(f"\nDimensiones del secador solar para α = {alpha}°:")
    print(f"{'ALTURA':<10}{'BASE':<10}{'ARISTA':<10}{'APOTEMA':<10}")
    print(f"{'-'*40}")
    for H in alturas:
        B, A, C = calcular_dimensiones(H, alpha)
        print(f"{H:<10.2f}{B:<10.2f}{A:<10.2f}{C:<10.2f}")

# Alturas para las cuales se calcularán las dimensiones
alturas = [0.5, 0.6, 0.7, 0.8, 0.9, 1.0, 1.1, 1.2, 1.3, 1.4, 1.5]

# Ángulos para los cuales se calcularán las dimensiones
angulos = [51.85, 43.0, 68.0]

# Mostrar las tablas de resultados para cada ángulo
for alpha in angulos:
    mostrar_tabla(alpha, alturas)
````
